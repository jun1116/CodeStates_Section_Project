```python
def timechange(x):
    date= x.split()[0].replace('.','-')
    y=date.split('-')[0]
    m=date.split('-')[1]
    d=date.split('-')[2]
    if len(m)==1:
        m = '0' + m
    if len(d)==1:
        d = '0' + d
    hour = x.split()[1].split(':')[0]
    if len(hour)==1 : 
        hour = '0'+hour
    return y + '-' + m + '-' + d + ' ' + hour
# df_gangnam.loc[:,'대여일시'].apply(lambda x : timechange(x))
```

```python
def gangcount(df_gangnam, year):
    대여카운트=pd.pivot_table(df_gangnam, index='대여일시', aggfunc='count')[['대여대여소번호','이용거리']]
    반납카운트=pd.pivot_table(df_gangnam, index='반납일시', aggfunc='count')[['반납대여소번호','이용거리']]
    # 대여카운트

    df_gang_counted=pd.DataFrame(pd.date_range(f'{year}-01-01', periods=24*365, freq='H'), columns=['대여일시'])
    df_gang_counted['대여일시'] = df_gang_counted.loc[:,'대여일시'].apply(lambda x : str(x)[:-6])
    # df_gang_counted
    df_gang_counted=pd.merge(df_gang_counted.loc[:], 대여카운트, on='대여일시',how='outer')
    df_gang_counted=pd.merge(df_gang_counted.loc[:], 반납카운트, left_on='대여일시',right_on='반납일시',how='outer')
    df_gang_counted.columns=['일시','대여량','d','반납량','dd']
    df_gang_counted.drop(['d','dd'],axis=1,inplace=True)
    df_gang_counted['일시']=df_gang_counted['일시'].astype('datetime64')
    df_gang_counted.fillna(0, inplace=True)
    df_gang_counted=df_gang_counted[df_gang_counted.일시!=0]
    return df_gang_counted

# 12/31 넘어간 데이터가 0으로 매핑되는것 ->  날리기 

df_gang2019_counted = gangcount(df_gangnam2019,'2019')
df_gang2020_counted = gangcount(df_gangnam2020,'2020')
```

# 평가지표
- 회귀: MAE, RMSE, R^2 등을 사용할 수 있습니다.

```python
### 이곳에서 과제를 진행해 주세요 ### 
def rmsle(y, y_,convertExp=True):
    if convertExp:
        y = np.exp(y),
        y_ = np.exp(y_)
    log1 = np.nan_to_num(np.array([np.log(v + 1) for v in y]))
    log2 = np.nan_to_num(np.array([np.log(v + 1) for v in y_]))
    calc = (log1 - log2) ** 2
    return np.sqrt(np.mean(calc))
```

RMSLE를 사용하겠습니다.  
RMSLE는 기존의 RMSE에 Log를 적용해준 지표이다.  
RMSLE는 이상치값에 덜민감하며, 상대적 error를 측정해주고, Under Estimation(적게예측한 경우)에 더 민감(더 큰 패널티 부여)합니다.  
대여량예측이 더 적을경우 값이 크게나온다는 뜻이며, 이는 대여량을 통하여 향후 시간대에 대한 따릉이 배치에 대한 평가지표로 적절하다고 생각됩니다.  
향후 log1p()로 변환된 값은, np.expm1()으로 원래의 스케일로 복원가능하다.
